# Loops
In this lecture we will see how loops work and what is loops.<br>
Looping is a sequence of instructions that is continually repeated until a certain condition is reached.<br>

# التكرار
في هذه المحاضرة سيتم توضيح معنى التكرار وكيفية التكرار في لغات البرمجة.<br>
التكرار في لغات البرمجة هو تكرار مجموعة من الأوامر بشكل متواصل إلى أن يتحقق شرط معين.

## Looping types
Generally speaking, there are 2 types of looping:<br>
1. Looping with known number of iteration and for this type of looping we usually use <code>for</code> statement.<br>
2. Looping until a certain condition if met (or unmet) and for this type of looping we use <code>while</code> statement.<br>

## أنواع التكرار
بشكل عام ، يوجد نوعان من التكرار:<br>
1. التكرار لعدد معين من المرات ولهذا النوع عادة نستعمل عبارة <code>for</code>.<br>
2. التكرار بوجود شرط معين ولهذا النوع نستعمل عبارة <code>while</code>.

### <code>for</code> loops
for loops in python takes a new value from its <strong>list</strong> of value on each iteration of the loop, hence it needs a placeholder variable to hold this value. This placeholder is called the loop counter. After the loop finished this placeholder value will be the last one called in the loop.

### التكرار باستخدام <code>for</code>
في لغة بايثون ، عندما تستخدم عبارة <code>for</code> للتكرار فإنه توجد قيمة لكل تكرار من التكرارات و تأخذ قيم هذه التكرارات من مجموعة من القيمة المرتبة و تسمى هذه المجموعة <code>list</code> ، ولما كانت تسعمل هذه المجموعة فإن التكرار يحتاج لمتغير ليحفظ هذه القيمة. يسمى هذه المتغير بعداد التكرار. بعد نهاية التكرار تكون قيمة المتغير اخر قيمة تم حفظها قبل نهاية التكرار.

```python
print("This is the first loop.")
for i in range(10): # Notice here the loop counter is called i
    print(i)
print("This is the end of the first loop.")
print(i) # Here i will be 9
print("This is the second loop.")
for i in range(1,11):
    print(i)
print("This is the end of the second loop.")
print(i) # Here i will be 10
print("This is the third loop.")
for i in range(0,10,2):
    print(i)
print("This is the end of the third loop.")
print(i) # Here i will be 8
```

```python
print("This is the fourth loop.")
for i in range(1,1):
    print(i)
print("This is the end of the fourth loop.")
print(i) # Here it will still be 8 since the loop didn't run
```

```python
a = [1,7,3,9,11] # This is list datatype in python
# Notice the use of [] and how the data are arranged inside
```

```python
# now we can loop throught the elements of the list a
for i in a:
    print(i)
    i = i + 5
    print(i)
# Notice here we changed the value of i inside the loop
print(i) # This will print the last element of the list a plus 5 so it will print 16
```

```python
print(a) # Now notice here we changed the
# value of the placeholder inside the loop 
# but notice the values of the list elements
# it didn't change
```

#### example
We gonna use the for loop to print the multiplication table of 7.

#### مثال
سوف نقوم باستعمال <code>for</code> لنخرج جدول ضرب للعدد 7

```python
# now let's remember we can define a variable
m = 7
# then we can print this variable
print(m)
```

```python
# now let’s remember we can do mathematical operation inside the print
print(7 * 8)
# now let's remeber we can do this using variable as well
print(m * 8)
```

```python
# now we can use this information to print the whole of the table of 7 like this
print("1 *", m, "=", 1 * m)
print("2 *", m, "=", 2 * m)
print("3 *", m, "=", 3 * m)
print("4 *", m, "=", 4 * m)
print("5 *", m, "=", 5 * m)
print("6 *", m, "=", 6 * m)
print("7 *", m, "=", 7 * m)
print("8 *", m, "=", 8 * m)
print("9 *", m, "=", 9 * m)
print("10 *", m, "=", 10 * m)
print("11 *", m, "=", 11 * m)
print("12 *", m, "=", 12 * m)
# this is too much code and writting
```

```python
# now let's do it using a for
for i in range(1, 13):
    print(i, "*", m, "=", i * m)
```
