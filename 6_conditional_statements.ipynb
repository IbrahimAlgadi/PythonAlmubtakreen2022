# Branching
Branching in programming is the changing of the normal flow of the execution of the code when special conditions are met.
There are 2 types of branching:<br>
A. Conditional.<br>
B. Loops

# التفرع
التفرع في البرمجة هو تغيير التنفيذ الطبيعي للبرنامج عندما تتحقق شروط معينة. يوجد نوعان للتفرع وهما:<br>
أ. التفرع الشرطي.<br>
ب. التكرار.

## Conditional  Branching
This type of branching interrupt the normal flow of execution the code line-by-line to exciting things in between or jumping lines to immediately to other lines to start executing.<br>The main statement for this type of branching is if (or if..else).<br>Whenever the compiler execute an if statement it checks if a certain condition is met or not

## التفرع الشرطي
هذا النوع من التفرع يقوم بتغيير التنفيذ الطبيعي للبرنامج من تنفيذ الأوامر بالترتيب إلى تغيير ترتيب هذا التنفيذ متى ما تحققت شروط معينة.<br>
الأمر الرئيسي لهذا التفرع هو if.<br>
عندما ينفذ المترجم أمر if يقوم بالتحقق من شرط معين مرفق مع العبارة.<br>

### The if statement
The if statement is a control statement that changes the flow of the execution of the code according to a certain condition with it, if the condition is the met the code inside the if body will be executed and when the code inside the if is finished executing the compiler will resume executing the code from after the if body, but when the condition isn't met the compiler will ignore the body of the if and start executing after it.

### عبارة if الشرطية
عبارة if هي عبارة تحكم تقوم بتغيير تنفيذ البرنامج اعتمادا على شرط معين، إذا تحقق الشرط يقوم المترجم بتنفيذ الأوامر بداخل الجملة الشرطية if وعندما ينتهي المترجم من تنفيذ هذه السلسلة من الأوامر يقوم المترجم بمواصلة تنفيذ الأوامر بشكل طبيعي بعد جسم if، إذا لم يتحقق الشرط يقوم المترجم بتجاهل الأوامر داخل جسم عبارة if ويواصل التنفيذ بعد الجسم مباشرة.

```python
x = 2
if (x == 2):
    print ("The variable x is two.")
print(x+5)
```

```python
x = 2
if (x == 1):
    print ("The variable x is one.")
print(x+5)
```

### The if else statement
The if else statement works the same way as the normal if with some differences. <br>1. We have 2 bodies one for the if and the other for the else body and the else is after the if like normal spoken language.<br>2. When the execution of the if body is finished the compiler will jump to the code after the else body.<br>3. If the condition isn’t met the compiler will execute the code inside the else body.
